- 1. MNIST digit image 전체 데이터를 모두 벡터로 변환하여 하나의 Numpy 행렬 X를 만든다.

```python
from sklearn.datasets import load_digits
import numpy as np
```

```python
digits = load_digits()
dimg = digits.images
X = dimg.reshape(1, 115008)
X = X.reshape(1797,64)
X
```

- 2. 모든 이미지의 조합에 대해 내적으로 유사도를 구해야 한다. 어떻게 구현하는 것이 효율적일까?

먼저 행렬 digits.images 데이터 전체를 행 벡터로 나열한 후 다시 reshape 과정을 거친 후 (1797 , 64) 행렬을 가진 X 변수에 저장합니다. 그 다음 내적을 통해 모든 이미지 조합에 대한 유사도를 구하면 (1797, 1797) 행렬의 정방행렬이 구해질 것 입니다.

# 연습문제 4

- 1. 위에서 만든 MNIST digit image 전체 데이터 행렬 X와 행렬 곱셈을 이용하여 모든 이미지의 조합에 대한 유사도를 구한다. 이 결과는 이미지의 데이터의 갯수를 크기로 하는 정방행렬이 된다.

```python
print(X.shape)
print("-----------------------------------------")
print((X.T).shape)
print("-----------------------------------------")
print(np.dot(X, X.T))
print("-----------------------------------------")
print(np.dot(X,X.T).shape)
```

- 2. 유사도 정방행렬을 이미지로 시각화 한다.

```python
import matplotlib.pylab as plt
%matplotlib inline
Y = np.dot(X,X.T); Y

f, ax = plt.subplots(1,1)
ax.imshow(Y, interpolation = 'nearest', cmap = plt.cm.bone_r)
ax.grid(False)
plt.show
```
