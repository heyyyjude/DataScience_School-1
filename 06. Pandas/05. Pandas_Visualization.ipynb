- 판다스를 통해서도 시각화를 할 수 있다.

```python
ts = pd.Series(np.random.randn(1000), index = pd.date_range('1/1/2000', periods = 1000))
ts = ts.cumsum() # 누적시킨다.
ts.plot()
plt.show()
```

```python
df = pd.DataFrame(np.random.randn(1000, 4), index = ts.index, columns = ['A','B','C','D'])
df = df.cumsum()
df.tail()
```

```python
df.plot()
plt.legend(loc='best')
plt.show()
```

```python
iris = sns.load_dataset("iris")
iris.tail()
```

```python
iris[:10].plot()
plt.show()
```

```python
iris[:10].plot(kind = "bar")
plt.show()
```

```python
iris_pie = iris.species.value_counts()
iris_pie.plot.pie(autopct = '%d%%') # decimal 표시
plt.axis('equal')
plt.show()
```

```python
iris.plot.scatter(x = "sepal_length", y = "sepal_width")
plt.show()
```

```python
setosa = iris[iris["species"] == "setosa"]
versicolor = iris[iris["species"] == "versicolor"]
virginica = iris[iris["species"] == "virginica"]
```

```python
ax = setosa.plot.scatter(x = 'sepal_length', y = 'sepal_width', c = 'red')
versicolor.plot.scatter(x = 'sepal_length', y = 'sepal_width', c = "blue" , ax = ax)
virginica.plot.scatter(x = "sepal_length", y = 'sepal_width', c = 'black', ax = ax)
plt.show()
```
