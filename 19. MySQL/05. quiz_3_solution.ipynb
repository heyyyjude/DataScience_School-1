```python
SQL_QUERY_VIEW = """
    create view language_use as
    select distinct(cl.language),

    round(country.population * cl.percentage * 0.01) as count

    from country
    join countrylanguage as cl
    on country.code = cl.countrycode;
"""
SQL_QUERY = """
    select language, sum(count) as count
    from language_use
    group by language
    order by count desc
    limit 10;
"""
```

```python
try:
    tmp_df = pd.read_sql(SQL_QUERY_VIEW, db)

except:
    quiz3_3_df = pd.read_sql_query(SQL_QUERY, db)
    quiz3_3_df['count'] = quiz3_3_df['count'].astype(int)
quiz3_3_df
```

```python
SQL_QUERY = """
    select city.Name, city.countrycode, country.name, round((city.population / country.population) * 100, 2) as percentage
    from country
    join city
    on country.code = city.countrycode
    where round(country.Population * 0.1) < city.population and city.population > 5000000
    order by percentage desc;
"""
```

```python
quiz3_4_df = pd.read_sql_query(SQL_QUERY, db)
quiz3_4_df
```

```python
SQL_QUERY_VIEW = """
    create view country_density as 
    select Code, Name, round(Population/SurfaceArea) as density
    from country
    where SurfaceArea > 10000 and round(Population/SurfaceArea) >= 200
    order by density desc;
"""
SQL_QUERY = """
    select cd.name, count(cd.name) as language_count
    from country_density as cd
    join countrylanguage as cl
    on cd.code = cl.countrycode
    group by cd.name
    having language_count > 5
    order by language_count desc;
"""
```

```python
try:
    tmp_df = pd.read_sql(SQL_QUERY_VIEW, db)

except:
    quiz3_5_df = pd.read_sql_query(SQL_QUERY, db)
    
quiz3_5_df
```

```python
SQL_QUERY_VIEW = """
    create view country_lang_three as
    select country.code, lang_count.name, lang_count.language_count, lang_count.language
    from country
    join 
    (select country.name, count(country.name) as language_count, group_concat(cl.language) as language
    from country
    join countrylanguage as cl
    on country.code = cl.countrycode
    group by country.name 
    having count(country.name) <= 3
    ) as lang_count
    on country.name = lang_count.name;
"""
SQL_QUERY = """
    select city.countrycode, city.name as city_name, city.population, clt.name, clt.language_count, clt.language
    from country_lang_three as clt
    join city
    on clt.code = city.countrycode
    where city.population > 3000000
    order by city.population desc;
"""
```

```python
try:
    tmp_df = pd.read_sql(SQL_QUERY_VIEW, db)

except:
    quiz3_6_df = pd.read_sql_query(SQL_QUERY, db)
    
quiz3_6_df
```
